Berikut merupakan penjabaran terkait informasi umum tiap fitur yang ada pada dataset. Berikut adalah penjelasannya.

1. **Count**: Adalah jumlah baris yang ada pada dataset. Terdapat 569 baris pada dataset ini.

2. **Mean**: Menjelaskan terkait nilai rata-rata yang dimiliki suatu fitur. Caranya, dengan menambahkan seluruh nilai pada masing-masing fitur akan dijumlahkan, kemudian dikurangi dengan total baris dataset. Contohnya, pada fitur radius_mean, dijumlahkan seluruhnya, didapat hasilnya 8038.429. Kemudian, hasilnya akan dibagi dengan jumlah baris, yakni 569. Sehingga didapat

> 8038.429 / 569 = 14.12729173989455

Artinya, nilai rata-rata dari fitur radius_mean adalah 14.12729173989455, yang bisa disederhanakan menjadi 1.412729e+01. Hal ini juga berlaku dengan fitur-fitur lainnya.

3. **std**: Std atau standar deviasi adalah seberapa tersebar atau seberapa jauh nilai-nilai dalam suatu dataset dari rata-rata (mean) dataset tersebut. Standar deviasi memberikan gambaran tentang sejauh mana titik data dari rata-rata, dan semakin tinggi standar deviasi, semakin besar variasi atau dispersi data.

Ada dua hal yang patut diperhatikan pada standar deviasi ini, yaitu:
*   Jika standar deviasi mendekati nol, maka sebagian besar nilai dalam dataset cenderung mendekati rata-rata.
*   jika standar deviasi besar dan tidak mendekati nol, maka nilai-nilai dalam dataset tersebar jauh dari rata-rata, dan dataset memiliki variasi yang tinggi.

Contohnya kita ambil radius_mean dan area_mean. Bila kita lihat berdasarkan tabel diatas, dapat disimpulkan bahwa radius_mean memiliki persebaran nilai yang lebih mendekati rata-rata dengan nilai **3.524049e+00** dibandingkan area_mean yang memiliki nilai **3.519141e+02**.

4. **min**: Menampilkan nilai minimum atau nilai terendah yang ada pada tiap fitur.

5. **25%**: Menampilkan nilai kuartil pertama yang ada di antara nilai min dan mean.

6. **50%**: Menampilkan nilai kuartil kedua (median) yang terletak di tengah-tengah data.

7. **75%**: Menampilkan nilai kuartil ketiga yang ada di antara median dan nilai maksimal suatu data.

8. **Max** = Menampilkan nilai maksimum yang ada pada tiap fiturnya.
